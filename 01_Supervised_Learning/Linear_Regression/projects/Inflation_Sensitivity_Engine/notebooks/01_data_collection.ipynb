```python
import pandas as pd
import numpy as np
from pathlib import Path

raw_path = Path("../data/raw")
raw_path.mkdir(parents=True, exist_ok=True)

print("Saving data to:", raw_path.resolve())

np.random.seed(42)
dates = pd.date_range("2015-01-01", "2023-12-01", freq="MS")

commodity_prices = pd.DataFrame({
    "date": dates,
    "commodity_price": np.random.normal(120, 10, len(dates)).cumsum()
})

fuel_prices = pd.DataFrame({
    "date": dates,
    "fuel_price": np.random.normal(70, 5, len(dates)).cumsum()
})

rainfall = pd.DataFrame({
    "date": dates,
    "rainfall_index": np.random.normal(100, 20, len(dates))
})

commodity_prices.to_csv(raw_path / "commodity_prices.csv", index=False)
fuel_prices.to_csv(raw_path / "fuel_prices.csv", index=False)
rainfall.to_csv(raw_path / "rainfall.csv", index=False)

print("\nFILES CREATED:")
for f in raw_path.iterdir():
    print("✔", f.name)
```
