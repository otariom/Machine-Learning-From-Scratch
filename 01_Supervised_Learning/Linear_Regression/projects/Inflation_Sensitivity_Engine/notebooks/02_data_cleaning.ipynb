```python
from pathlib import Path

required_files = [
    "../data/raw/commodity_prices.csv",
    "../data/raw/fuel_prices.csv",
    "../data/raw/rainfall.csv"
]

for file in required_files:
    assert Path(file).exists(), f"Missing file: {file}. Run data_collection first."
```

```python
import pandas as pd

commodity_prices = pd.read_csv("../data/raw/commodity_prices.csv", parse_dates=["date"])
fuel_prices = pd.read_csv("../data/raw/fuel_prices.csv", parse_dates=["date"])
rainfall = pd.read_csv("../data/raw/rainfall.csv", parse_dates=["date"])

df = commodity_prices.merge(fuel_prices, on="date", how="inner")
df = df.merge(rainfall, on="date", how="inner")

df = df.sort_values("date")
df = df.fillna(method="ffill")

df.to_csv("../data/processed/merged_data.csv", index=False)
df.head()
```
